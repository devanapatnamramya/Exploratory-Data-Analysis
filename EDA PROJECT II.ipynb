### Find out the distribution of sales over different times of the day (Morning, Afternoon, Evening, Night) and answer the following for each city;

```python
def f(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'After Noon'
    elif (x > 16) and (x <= 20) :
        return 'Evening'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'
```

```python
df1['session'] = df1['Hour'].apply(f)
```

```python
df1
```

### What products are mostly sold at different times of the day

```python
Day_sales = df1.groupby(['Product','session'],as_index=False).sum().sort_values(by ='Sales',ascending=False)
```

```python
Day_sales
```

```python
plt.figure(figsize=(20, 18))
ax = sns.barplot(y = 'Product',x= 'Sales',hue = 'session',data =Day_sales )
```

### By this we can conclude that we have each and every product has high sales in evening between 4 pm to 8 pm

### Which time of day has the most sales

```python
session1= df1.groupby(['session']).agg({'Sales':'count'}).reset_index()
```

```python
session1
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="session",y ="Sales",hue ='session', data=session1,dodge=False)
```

### Evening time has highest sales

### Based on analysis, Evening time(4PM to 8PM) is the best working hours and the best products are Macpro laptop and iphone for the retail stores in each city
